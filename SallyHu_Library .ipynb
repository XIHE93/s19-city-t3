```python
## import pandas
import os
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## read "UW Data Science Project" csv file into data frame
madisonFinancedf = pd.read_csv("/Users/HuKexin/desktop/s19-city-t3-master/spend.csv")
```

```python
##subset data by agency, account tye and year
subsetAllAgency = madisonFinancedf[["Agency","Account Type","2015","2016","2017","2018"]]
```

```python
##subset data by agency(library)
criteria = subsetAllAgency["Agency"]=="50 -  LIBRARY"
libraryData = subsetAllAgency[criteria]
```

```python
##subset library data into revenue and group by year
criteria_account_type_revenue = libraryData["Account Type"]=="Revenue"
revenue_libraryData = libraryData[criteria_account_type_revenue]
##year 2015
revenue_2015 = revenue_libraryData["2015"].fillna(0)
sum_revenue_2015 = sum(abs(revenue_2015))
##year 2016
revenue_2016 = revenue_libraryData["2016"].fillna(0)
sum_revenue_2016 = sum(abs(revenue_2016))
##year 2017
revenue_2017 = revenue_libraryData["2017"].fillna(0)
sum_revenue_2017 = sum(abs(revenue_2017))
##year 2018
revenue_2018 = revenue_libraryData["2018"].fillna(0)
sum_revenue_2018 = sum(abs(revenue_2018))
```

```python
##subset library data into expense and group by year
criteria_account_type_expense = libraryData["Account Type"]=="Expense"
expense_libraryData = libraryData[criteria_account_type_expense] 
##year 2015
expense_2015 = expense_libraryData["2015"].fillna(0)
sum_expense_2015 = sum(expense_2015)
##year 2016
expense_2016 = expense_libraryData["2016"].fillna(0)
sum_expense_2016 = sum(expense_2016)
##year 2017
expense_2017 = expense_libraryData["2017"].fillna(0)
sum_expense_2017 = sum(expense_2017)
##year 2018
expense_2018 = expense_libraryData["2018"].fillna(0)
sum_expense_2018 = sum(expense_2018)
```

```python
##Graph 1: Compare growing trend of revenue and expense during 2015-2018
##create a new dataframe to store subset data and plot graph
index = ["Revenue","Expense"]
year_2015 = [sum_revenue_2015,sum_expense_2015]
year_2016 = [sum_revenue_2016,sum_expense_2016]
year_2017 = [sum_revenue_2017,sum_expense_2017]
year_2018 = [sum_revenue_2018,sum_expense_2018]
df_index_account_type = pd.DataFrame({"2015":year_2015,"2016":year_2016,"2017":year_2017,"2018":year_2018},index = index)
ax = df_index_account_type.plot.bar()
plt.ylabel("Amount of Money($)")
plt.xlabel("Account Type")
plt.title("Trend of Expenditure and Expense during 2015-2018")

##Discussion: 
##Both revenue and expense are increasing each year
```

```python
##subset library data into year and group by account type(revenue)
##year 2015
year_2015_1 = libraryData[["Account Type","2015"]]
year_2015_1 = year_2015_1.fillna(0)
criteria_year_2015_revenue = year_2015_1["Account Type"]=="Revenue"
year_2015_revenue = year_2015_1[criteria_year_2015_revenue]
sum_2015_revenue = sum(abs(year_2015_revenue["2015"]))
##year 2016
year_2016_1 = libraryData[["Account Type","2016"]]
year_2016_1 = year_2016_1.fillna(0)
criteria_year_2016_revenue = year_2016_1["Account Type"]=="Revenue"
year_2016_revenue = year_2016_1[criteria_year_2016_revenue]
sum_2016_revenue = sum(abs(year_2016_revenue["2016"]))
##year 2017
year_2017_1 = libraryData[["Account Type","2017"]]
year_2017_1 = year_2017_1.fillna(0)
criteria_year_2017_revenue = year_2017_1["Account Type"]=="Revenue"
year_2017_revenue = year_2017_1[criteria_year_2017_revenue]
sum_2017_revenue = sum(abs(year_2017_revenue["2017"]))
##year 2018
year_2018_1 = libraryData[["Account Type","2018"]]
year_2018_1 = year_2018_1.fillna(0)
criteria_year_2018_revenue = year_2018_1["Account Type"]=="Revenue"
year_2018_revenue = year_2018_1[criteria_year_2018_revenue]
sum_2018_revenue = sum(abs(year_2018_revenue["2018"]))
```

```python
##subset library data into year and group by account type(expense)
##year 2015
year_2015_1 = libraryData[["Account Type","2015"]]
year_2015_1 = year_2015_1.fillna(0)
criteria_year_2015_expense = year_2015_1["Account Type"]=="Expense"
year_2015_expense = year_2015_1[criteria_year_2015_expense]
sum_2015_expense = sum(year_2015_expense["2015"])
##year 2016
year_2016_1 = libraryData[["Account Type","2016"]]
year_2016_1 = year_2016_1.fillna(0)
criteria_year_2016_expense = year_2016_1["Account Type"]=="Expense"
year_2016_expense = year_2016_1[criteria_year_2016_expense]
sum_2016_expense = sum(year_2016_expense["2016"])
##year 2017
year_2017_1 = libraryData[["Account Type","2017"]]
year_2017_1 = year_2017_1.fillna(0)
criteria_year_2017_expense = year_2017_1["Account Type"]=="Expense"
year_2017_expense = year_2017_1[criteria_year_2017_expense]
sum_2017_expense = sum(year_2017_expense["2017"])
##year 2018
year_2018_1 = libraryData[["Account Type","2018"]]
year_2018_1 = year_2018_1.fillna(0)
criteria_year_2018_expense = year_2018_1["Account Type"]=="Expense"
year_2018_expense = year_2018_1[criteria_year_2018_expense]
sum_2018_expense = sum(year_2018_expense["2018"])
```

```python
##Graph 2: Compare difference of expense and revenue in each of these four years
##create a new dataframe to store subset data and plot graph
index = ["2015","2016","2017","2018"]
total_expenditure = [sum_2015_expense,sum_2016_expense,sum_2017_expense,sum_2018_expense]
total_revenue = [sum_2015_revenue,sum_2016_revenue,sum_2017_revenue,sum_2018_revenue]
df_index_account_type = pd.DataFrame({"Expense":total_expenditure,"Revenue":total_revenue},index = index)
ax = df_index_account_type.plot.bar()
plt.xlabel("year")
plt.ylabel("amount of money($)")
plt.title("Difference between Expenditure and Expense in Each Year(2015-2018)")

##Discussion:
##Expense and revenue are almost balance during these four years. 
##Actually, in year 2015 and 2017, revenue exceeds expense for a little bit. 
                    
```

```python
##Graph 3: Trend of expenditure during 2015-2018
##line plot(expense)
x =["2015","2016","2017","2018"]
y_expense = total_expenditure
y_expense = pd.Series(y_expense)
plt.plot(x,y_expense)
plt.xlabel("year")
plt.ylabel("Amount of Money($ 1 billion)")
plt.title("Trend of Expenditure during 2015-2018")
plt.show()
##discussion:
##expenditure rises every year and the increasing number is nealy a linear straight line
```

```python
##Graph 4: Trend of revenue during 2015-2018
##line plot(revenue)
x =["2015","2016","2017","2018"]
y_revenue = total_revenue
y_revenue = pd.Series(y_revenue)
plt.plot(x,(y_revenue/(10^9)))
plt.xlabel("year")
plt.ylabel("Amount of Money($ 1 billion)")
plt.title("Trend of Revenue during 2015-2018")
plt.show()
##discussion: 
##revenue increases each year and there is a big increase between year 2016 and 2017(slope is steeper)
```

```python
##Graph 5: combine expenditure and revenue line chart into one graph
plt.plot(x,y_expense, label = "Expense")
plt.plot(x,y_revenue,label = "Revenue")
plt.legend(loc='upper left')
plt.title("Expense and Expenditure during 2015-2018")
plt.xlabel("Year")
plt.ylabel("Amount of Money($ 10 million)")
```
