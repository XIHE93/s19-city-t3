```python
# graph for 2018 cd
# https://github.com/tylerharter/caraza-harter-com/blob/master/tyler/cs301/fall18/materials/code/lec-33/top-directors.ipynb
groups = {}
for t in madison_community_development[madison_community_development.notnull()].iterrows():
    row = t[1]
    org = row['Organization']
    if not org in groups:
        groups[org] = []
    groups[org].append(row)

d = {}
other_sum= 0
for org in groups:
    rows = groups[org]
    budget_sum = 0
    for row in rows:
        budget_sum += row['2018']
    if budget_sum < 50000:
        other_sum += budget_sum
    else:
        d[org]=budget_sum

d["Other"] = other_sum
    
    

graph_2018_community= pd.DataFrame.from_dict(d, orient='index', columns=['2018 budget for community deve ($1000)'])

```

```python
ax = graph_2018_community.plot.bar()
ax.set_ylabel('2018 budget for community development ')
```

```python
# https://github.com/tylerharter/caraza-harter-com/blob/master/tyler/cs301/fall18/materials/code/lec-33/top-directors.ipynb
bud = {}
for t in madison_community_development.iterrows():
    row = t[1]
    org = row['Organization']
    if not org in bud:
        bud[org] = []
    bud[org].append(row)

f = {}
other_sum= 0
for org in bud:
    rows = bud[org]
    budget_sum_2015 = 0
    budget_sum_2016 = 0
    budget_sum_2017 = 0
    budget_sum_2018 = 0
    other_sum = 0
    for row in rows:
        if budget_sum_2015 < 5000:
            other_sum += budget_sum_2015
        elif pd.notnull(row["2015"]):
            budget_sum_2015 += row['2015']
        if budget_sum_2015 < 5000:
            other_sum += budget_sum_2016
        elif pd.notnull(row["2016"]):
            budget_sum_2016 += row['2016']
        if budget_sum_2015 < 5000:
            other_sum += budget_sum_2017
   
        elif pd.notnull(row["2017"]):
            budget_sum_2017 += row['2017']
        if budget_sum_2015 < 5000:
            other_sum += budget_sum_2018

        elif pd.notnull(row["2018"]):
            budget_sum_2018 += row['2018']
        
    f[org]= [budget_sum_2015, budget_sum_2016, budget_sum_2017, budget_sum_2018, other_sum]
    
f= pd.DataFrame.from_dict(f, orient='index', columns=['2015', '2016', '2017', '2018', 'other'])
```

```python
ax = f.plot.bar()
ax.set_ylabel(' budget for community development ')
```
