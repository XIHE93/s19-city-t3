```python
spendframe=pd.read_csv("spend.csv")
```

```python
EngineeringFrame=spendframe[spendframe["ID"]==40]
```

```python
EngineeringFrame= EngineeringFrame.drop('ID',1)
```

```python
EngineeringSpending = EngineeringFrame[EngineeringFrame["Account Type"]=="Expense"]
EngineeringRevenue = EngineeringFrame[EngineeringFrame["Account Type"]=="Revenue"]
```

```python
EngineeringSpending.describe()
```

```python
EngineeringRevenue.describe()
```

```python
#Spending by type
spending_type = EngineeringSpending.groupby("Code")
```

```python
for code, code_df in spending_type:
    print(code)
    print(code_df)
```

```python
spendingsummary = spending_type.sum()
spendingsummary.plot.bar()
plt.title('Spending VS. Categories')
plt.xlabel('Categories')
plt.ylabel('Spending')
```

```python
revenue_type = EngineeringRevenue.groupby("Code")
```

```python
revenuesummary = revenue_type.sum()
revenuesummary.plot.bar()
```
